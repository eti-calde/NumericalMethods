### Rank-1 matrices

$$
A = uv^{T} = 
\begin{bmatrix}
-  & u_{0}v^{T} & -\\
-  & u_{1}v^{T} & -\\
& \vdots &\\
-  & u_{n}v^{T} & -\\
\end{bmatrix} = 
\begin{bmatrix}
| & | &  & |\\
v_0u & v_2u & \dots & v_nu\\
| & | &  & |\\

\end{bmatrix}

$$

```python
u = np.array([1,2,3,4]).T
# v = np.eye(3) + np.eye(3, k=1)
v = np.array([1,1,1])

```

```python
np.vstack([
u[0,:] @ v.T,
u[1,:] @ v.T,
u[2,:] @ v.T
])
```
